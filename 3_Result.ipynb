```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split 

import pickle

plt.rcParams.update({
    'font.family':'Times New Roman', 
    'font.size': 14, 
    'axes.titlesize': 16, 
    'axes.labelsize': 14, 
    'xtick.labelsize': 12, 
    'ytick.labelsize': 12,
    'legend.fontsize': 14, 
})
```

# 1.分析损失

```python
with open('experiment_1_loss.pkl', 'rb') as f:
    experiment_1_loss = pickle.load(f)
```

```python
# experiment_1_loss_mean = []
# experiment_1_loss_std = []
experiment_1_loss_median = []
experiment_1_loss_min = []


for sublist in experiment_1_loss:
    # mean_value = np.mean(sublist[-50:]) 
    # std_dev_value = np.std(sublist[-50:]) 
    median = np.median(sublist[-50:])
    min_v = np.min(sublist[-50:])
    # experiment_1_loss_mean.append(mean_value)
    # experiment_1_loss_std.append(std_dev_value)
    experiment_1_loss_median.append(median)
    experiment_1_loss_min.append(min_v)
```

```python
train_median = experiment_1_loss_median[::2]
val_median = experiment_1_loss_median[1::2]
train_min = experiment_1_loss_min[::2]
val_min = experiment_1_loss_min[1::2]

experiment_labels = [f'1-{i+1}' for i in range(10)]

x = np.arange(len(experiment_labels))

width = 0.2

fig, ax = plt.subplots(figsize=(10, 6))

rects1 = ax.bar(x - 1.5 * width, train_median, width, label='Train Median')
rects2 = ax.bar(x - 0.5 * width, val_median, width, label='Val Median')
rects3 = ax.bar(x + 0.5 * width, train_min, width, label='Train Min')
rects4 = ax.bar(x + 1.5 * width, val_min, width, label='Val Min')

ax.set_xlabel('Experiment')
ax.set_ylabel('Loss')
ax.set_title('Training and Validation Loss Statistics for Experiments 1-1 to 1-10')
ax.set_xticks(x)
ax.set_xticklabels(experiment_labels)

ax.legend()

ax.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('experiment1 损失分析.svg', dpi=300) 
plt.show()
```
